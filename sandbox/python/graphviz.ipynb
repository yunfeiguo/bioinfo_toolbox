```python
class UndirectedGraph():
    def __init__(self, input):
        '''read undirected edges from input file
        and construct an undirected graph
        
        assume each has 2 words, separated by whitespace
        nodeA nodeB
        '''
        self.nodes = {}
        self.edges = {}
        for l in input:
            f = l.split()
            assert len(f) == 2
            nodeA = self.nodes.setdefault(f[0], self.Node(f[0]))
            nodeB = self.nodes.setdefault(f[1], self.Node(f[1]))
            self.edges.setdefault(nodeA, []).append(nodeB)
            self.edges.setdefault(nodeB, []).append(nodeA)
    class Node:
        def __init__(self, label):
            self.label = label
        def __str__(self):
            return self.label
    def draw(self):
        '''
        return dot representation of 
        '''
        import graphviz
        g = graphviz.Graph(comment = 'test')
        for node in self.edges.keys():
            g.node(str(node), str(node))
        for src, dsts in iter(self.edges.items()):
            for dst in dsts:
                g.edge(str(src), str(dst))
        return g
class DirectedGraph():
    def __init__(self, input):
        '''read directed edges from input file
        and construct an directed graph
        
        assume each has 2 words, separated by whitespace
        parent child
        '''
        self.nodes = {}
        self.edges = {}
        for l in input:
            f = l.split()
            assert len(f) == 2
            nodeA = self.nodes.setdefault(f[0], self.Node(f[0]))
            nodeB = self.nodes.setdefault(f[1], self.Node(f[1]))
            self.edges.setdefault(nodeA, []).append(nodeB)
    class Node:
        def __init__(self, label):
            self.label = label
        def __str__(self):
            return self.label
    def draw(self):
        '''
        return dot representation of 
        '''
        import graphviz
        g = graphviz.Digraph(comment = 'test')
        for node in self.edges.keys():
            g.node(str(node), str(node))
        for src, dsts in iter(self.edges.items()):
            for dst in dsts:
                g.edge(str(src), str(dst))
        return g
```
